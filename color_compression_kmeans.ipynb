**Color Compression**


Color compression is an interesting application of clustering. 

We will use a sample image from the SciKit-Learn data set.

```
from sklearn.datasets import load_sample_image
import matplotlib.pyplot as plt

china = load_sample_image("china.jpg")
ax = plt.axes(xticks = [], yticks = [])
ax.imshow(china);
```

Each image is made of up pixels, and each pixel is represented by 3 numbers indicting a value for red, green, and blue colors.

The sample image is stored in a three-dimensional array of size (height, width, RGB), containing red/blue/green contributions as integers from 0 to 255.

```
china.shape
```

Height: 427 pixels; Width: 640 pixels; RGB: 3 dimensions.

Since each value RGB value is between 0 and 255, we can take the percentage of each value and align all the pixels in one row.

```
data = china / 255.0 # use 0 ... 1 scale
data = data.reshape(427 * 640, 3)
data.shape
```

We will use K-means clustering to cluster similar colors into one one color. Since we want to produce an image just using 16 colors, we will 16 clusters.

```
from sklearn.cluster import MiniBatchKMeans
kmeans = MiniBatchKMeans(16)
kmeans.fit(data)
new_colors = kmeans.cluster_centers_[kmeans.predict(data)]
```

We can now display the two images side by side to compare them.

```
china_recolored = new_colors.reshape(china.shape)

fig, ax = plt.subplots(1, 2, figsize = (16, 6), 
                      subplot_kw=dict(xticks=[], yticks=[]))
fig.subplots_adjust(wspace=0.05)
ax[0].imshow(china)
ax[0].set_title('Original Image', size = 16)
ax[1].imshow(china_recolored)
ax[1].set_title('16-color Image', size = 16)
```
